# Text to speech conversion using gtts

```python
# pip install gtts
```

```python
from gtts import gTTS
from IPython.display import Audio
```

```python
def text2speech1():
    text =  '''Murphy’s Law means: if something can go wrong, it probably will.
               Like if you carry an ice cream cone too fast, you’ll probably drop it. Or if you leave your toy outside, it’ll rain.
               It’s not magic or bad luck—it’s just a way to remind us to be careful and think ahead. It’s kind of like the universe saying, 
               ‘Let’s see what happens if you’re not paying attention.’ So if you double-check stuff, you can outsmart Murphy'''
    
    # You can change 'hi' to 'en' for English, or another supported language
    # tld='com' sets the accent (e.g., 'co.in', 'co.uk', etc.).
    tts = gTTS(text, lang='en', tld='co.uk')  
    tts.save('text_to_speech_gtts.wav')

    return Audio('text_to_speech_gtts.wav', autoplay=False)
```

```python
# Run the function
text2speech1()
```

# Text to speech conversion using pyttsx3

```python
# pip install pyttsx3
```

```python
import pyttsx3
from IPython.display import Audio
```

```python
def text2speech2():
    text = '''Some people believe that things like ‘good’ and ‘bad’ aren’t real rules that exist in the world. They’re just ideas people made up to help 
              them live together. That’s called moral nihilism. Imagine if there were no rules at all—no one saying what’s right or wrong. A moral 
              nihilist says: there’s no magic list in the sky saying what’s good or bad. It’s just people deciding that stuff together. But even if you
              don’t believe in right or wrong rules, you can still care about people and try to help. You don’t need invisible rules to be kind.'''
    
    audio = pyttsx3.init()
    audio.setProperty('rate', 150)
    audio.setProperty('volume', 0.8)
    
    # Change the voices
    voice = audio.getProperty('voices')
    
    # 0 for male and 1 for female voice
    audio.setProperty('voice', voice[0].id)
    
    # text-to speech conversion
    audio.say(text)

    # save the audio file
    audio.save_to_file(text, 'test_male_Voice.mp3')
    #audio.save_to_file(text, 'test_female_Voice.mp3')

    return audio.runAndWait()
```

```python
text2speech2()
```
